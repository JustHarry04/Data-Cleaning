```python
clean1= clean1.drop(columns=["Unnamed: 0"])
```

```python
clean1.info()
```

```python
clean1['Safety'] = clean1['Safety'].replace({0: 'Not_safe', 1: 'safe'})
```

```python
for_columns=[]
for column in clean1.columns:
    for_columns.append(column)

```

```python

for column in for_columns:
    if clean1[column].dtype == np.int64 or clean1[column].dtype == np.float64:
        clean1[column] = clean1[column].fillna(clean1[column].median())
    else:
        if not clean1[column].mode().empty:
            clean1[column] = clean1[column].fillna(clean1[column].mode().iloc[0])
        else:
            clean1[column] = clean1[column].fillna("")

for column in for_columns:
    if clean1[column].dtype in [np.int64, np.float64]:
        clean1[column] = clean1[column].fillna(clean1[column].median())
    else:
        mode_value = clean1[column].mode()
        if not mode_value.empty:
            clean1[column] = clean1[column].fillna(mode_value.iloc[0])
        else:
            clean1[column] = clean1[column].fillna("")

```
